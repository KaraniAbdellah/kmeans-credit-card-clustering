```python
# Import Packages
from sklearn.cluster import KMeans
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from matplotlib import pyplot as plt
```

```python
# Load CSV File
csv_file = pd.read_csv("./income.csv")
csv_file.all() # Attributs
csv_file.head()
```

```python
# Draw Two Axes
plt.scatter(csv_file["Age"], csv_file["Income($)"])

# Scalling The Values
scaler = MinMaxScaler()
# Find Min, Max Value for Income Column
scaler.fit(csv_file[["Income($)"]])
csv_file[["Income($)"]] = scaler.transform(csv_file[["Income($)"]])

# Find Min and Max Values for Age Column
scaler.fit(csv_file[["Age"]])
csv_file[["Age"]] = scaler.transform(csv_file[["Age"]])

# Display The DataSet
csv_file
```

```python
# Define Clusters With K-Means
km = KMeans(n_clusters=2)

# Chose Centroid, Calcute Distance, and Get Cluster for Each Point
y_predicted = km.fit_predict(csv_file[["Age", "Income($)"]])
y_predicted

# Assign for Each Row a Cluster
csv_file["cluster"] = y_predicted
csv_file

# Draw It For Sew It Good
csv_file.cluster
c0 = csv_file[csv_file.cluster == 0]
c1 = csv_file[csv_file.cluster == 1]
plt.scatter(c0["Age"], c0["Income($)"], color="red")
plt.scatter(c1["Age"], c1["Income($)"], color="green")

# Get Centroid
km.cluster_centers_

# Draw Centroid
plt.scatter(km.cluster_centers_[:,0], km.cluster_centers_[:, 1])
```
