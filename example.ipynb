```python
# Import Packages
from sklearn.cluster import KMeans
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from matplotlib import pyplot as plt
```

```python
# Load The CSV FILE
df = pd.read_csv("./income.csv")
df.head()
```

```python
# Draw Two Axes (Age and Income)
plt.scatter(df['Age'], df["Income($)"])
```

```python
# Create the K-Means (Blan Algorithm)
km = KMeans(n_clusters=2)
km
```

```python
# First Epoch for Find Clusters
# We Have Two Clusters So We Presented as (0, 1)
y_predicted = km.fit_predict(df[["Age", "Income($)"]])
y_predicted

# Display Clusters for Each Row
df["cluster"] = y_predicted
df.head()
```

```python
df1 = df[df.cluster == 0]
df2 = df[df.cluster == 1]
plt.scatter(df1["Age"], df1["Income($)"], color="red")
plt.scatter(df2["Age"], df2["Income($)"], color="green")
```

```python
# shrinks the valus of income to the values between 0 and 1
scaler = MinMaxScaler()
scaler.fit(df[["Income($)"]])
df[["Income($)"]] = scaler.transform(df[["Income($)"]])
df

scaler.fit(df[["Age"]])
df[["Age"]] = scaler.transform(df[["Age"]])
df
```

```python
# Let Get The Cluster Knwo After Scaling the values
km = KMeans(n_clusters=2)
y_predicted = km.fit_predict(df[['Age', 'Income($)']])
y_predicted

df.drop('cluster', axis="columns", inplace=True)
df["cluster"] = y_predicted
df

df1 = df[df.cluster == 0]
df2 = df[df.cluster == 1]
plt.scatter(df1["Age"], df1["Income($)"], color="red")
plt.scatter(df2["Age"], df2["Income($)"], color="green")

# Get Centroid
km.cluster_centers_
plt.scatter(km.cluster_centers_[:,0], km.cluster_centers_[:,1])
```
